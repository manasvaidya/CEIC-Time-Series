### Create API Scraping function

```python
def ceic_series_scrape(ceic_result):
    # Initialise master df
    df_master = pd.DataFrame()
    
    # Set number of series scraped
    n = len(ceic_result.data)
    
    # Loop to extract data from all series
    for i in range(n):
        # Initialise empty df for each series
        df_series = pd.DataFrame()
        
        # Find length of i'th time series
        m=len(ceic_result.data[i].series_data.time_points)
        
        # Find Series i's name
        series_name = ceic_result.data[i].series_data.metadata.name
        
        # Loop to extract all time_points
        for j in range(m):
            df_series.loc[j,'timestamp'] = ceic_result.data[i].series_data.time_points[j].date
            df_series.loc[j, series_name] = ceic_result.data[i].series_data.time_points[j].value
            
        # Change Timestamp column to index for the series
        df_series = df_series.set_index('timestamp')
        
        # Merge with master dataframe
        df_master = pd.concat([df_master, df_series], axis=1)
        
    #return master df
    return df_master
            
            
```

### Run scrape

```python
ceic_series_scrape(ceic_df)
```
